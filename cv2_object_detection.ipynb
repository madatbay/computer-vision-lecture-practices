````python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import imutils
from imutils.object_detection import non_max_suppression
````

````python
im = plt.imread("/Users/madatbay/Development/cv/people.jpg")
````

````python
plt.imshow(im)
````

### Create HOG descriptor for person

````python
hog = cv2.HOGDescriptor()
hog.setSVMDetector(cv2.HOGDescriptor_getDefaultPeopleDetector())
````

### Identify rectangle of detected objects

````python
(r,w) = hog.detectMultiScale(im, winStride=(2,2), padding=(2,2), scale=1.1)
````

````python
rects = np.array([[x, y, x+w, y+h] for (x, y, w, h) in r])
````

````python
rects_NMS = non_max_suppression(rects, probs=None, overlapThresh=0.8)
````

````python
for (x1, y1, x2, y2) in rects_NMS:
    cv2.rectangle(im, (x1, y1), (x2, y2), (0, 255, 0), 1)
````
